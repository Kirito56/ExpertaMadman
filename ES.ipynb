```python
def start(engine: int,
          delfacts: bool = False,
          numoffact: bool = False,
          modify: bool = False,
          clear: bool = False,
          rules: bool = False):
    if engine == 1:
        e = DefFact()
        e.reset()
        e.init_card()
        print(e.facts)
        if delfacts and numoffact:
            num = int(input('Fact Number: '))
            e.retract(numoffact)
        elif clear:
            e.facts.clear()
        elif rules:
            e.run()
    elif engine == 2:
        e = Asserts()
        e.reset()
        e.init_graphic_card()
        print(e.facts)
        if delfacts:
            e.retract(numoffact)
```

```python
engine = int(input(f'Engine:\n1.\tDefFacts\n2.\tAsserts\nChoice:\t{int}\t-\t'))
delete = bool(input(f'Delete\tFact:\nEmpty.\tNo\n1.\tYes\nChoice:\t{int}\t-\t'))
clear = bool(input(f'Clear:\nEmpty.\tNo\n1.\tYes\nChoice:\t{int}\t-\t'))
rules = bool(input(f'Rules:\nEmpty.\tNo\n1.\tYes\nChoice:\t{int}\t-\t'))
print(f'Engine:\t{engine}\nDelete:\t{delete}\nClear:\t{clear}\nRules:\t{rules}')
if delete or clear or rules:
    numoffact = True
    start(engine, numoffact=numoffact, delfacts=delete, clear=clear, rules=rules)
```
